```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/Users/analytical/Desktop/GFSC/Projects/Rice Authentication Project/Rice-authentication-ICP-MS/grand.csv")
```

```python
X = data.drop('lv', axis = 1).values
y = data.lv.values
print(len(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state= 234)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import classification_report
```

```python
from skrebate import ReliefF
relief= ReliefF(n_jobs=-1, n_neighbors = 100)
relief.fit(X, y)
```

```python
feature_indecies = list(relief.top_features_)
```

```python
print(feature_indecies)
```

```python
result = pd.DataFrame(columns=['Elements','Importance'])
for feature_name, feature_score in zip(data.drop('lv', axis=1).columns, relief.feature_importances_): 
    result = result.append({'Elements': feature_name, 'Importance': feature_score}, ignore_index=True)
result = result.sort_values('Importance', ascending=False)
print(result)
```

```python
from sklearn.feature_selection import RFE
```

```python
# rfe feature selection
model = RandomForestClassifier(n_estimators=500, n_jobs=-1)
rfe = RFE(model)
fit = rfe.fit(X, y)
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier(n_estimators=500, n_jobs=-1)
model.fit(X, y)
print(model.feature_importances_)
```

```python
rf_params = {'n_estimators':[100,500], 'max_features':[1,2,3]}
```

```python
result = pd.DataFrame(columns=['Elements','Importance'])
for feature_name, feature_score in zip(data.drop('lv', axis=1).columns, model.feature_importances_): 
    result = result.append({'Elements': feature_name, 'Importance': feature_score}, ignore_index=True)
result = result.sort_values('Importance', ascending=False)
print(result)
```

```python
# using only top 3
X_selected = data[['Na','Al','Cd','Rb','Ba']]
Xs_train, Xs_test, ys_train, ys_test = train_test_split(X_selected, y, test_size = 0.4, random_state= 233)
gs_rf = GridSearchCV(estimator = RandomForestClassifier(), param_grid= rf_params, cv = 10, n_jobs=-1) 
gs_rf.fit(X_train, y_train) 
best_params = gs_rf.best_params_
best_rf = RandomForestClassifier(n_estimators= best_params["n_estimators"], max_features=best_params["max_features"])
best_rf.fit(Xs_train, ys_train)

ys_pred = best_rf.predict(Xs_test) 
print(classification_report(ys_test, ys_pred))
```
